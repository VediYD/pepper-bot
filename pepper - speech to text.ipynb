```python
import speech_recognition as sr

def convert_wav_to_text(audio_path):
    r = sr.Recognizer()

    # Load audio file
    with sr.WavFile(audio_path) as source:
        audio = r.record(source)

    # Convert audio to text
    try:
        text = r.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        print("Speech recognition could not understand audio")
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))

# Provide the path to your WAV file
wav_file_path = "recordings/file.wav"
result = convert_wav_to_text(wav_file_path)

# Print the converted text
print(result)
```

```python
import speech_recognition as sr

def convert_audio_to_text(audio_file):
    # Create a recognizer object
    r = sr.Recognizer()

    # Load audio file
    with sr.AudioFile(audio_file) as source:
        # Read the entire audio file
        audio = r.record(source)

    try:
        # Use the Google Speech Recognition engine to perform speech recognition
        text = r.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        print("Speech recognition could not understand audio")
    except sr.RequestError as e:
        print("Could not request results from speech recognition service: {}".format(e))

# Example usage
audio_file = "recordings/audio.wav"
text = convert_audio_to_text(audio_file)
print(text)
```

### Deep Speech Model

```python
pip install deepspeech
```

```python
import deepspeech

# Path to the DeepSpeech pre-trained model
model_path = 'deepspeech-0.9.3-models.pbmm'
# Path to the DeepSpeech pre-trained scorer 
scorer_path = 'deepspeech-0.9.3-models.scorer'
# Path to the WAV audio file you want to transcribe
audio_path = 'sample.wav'

# Create a DeepSpeech instance with the provided model
deepspeech_model = deepspeech.Model(model_path)

# Add scorer for improved accuracy if a scorer path is provided
if scorer_path:
    deepspeech_model.enableExternalScorer(scorer_path)

# Read the audio file
with open(audio_path, 'rb') as audio_file:
    audio_data = audio_file.read()

# Perform the speech recognition
text = deepspeech_model.stt(audio_data)
print(text)
```

```python
import speech_recognition as sr

def convert_audio_to_text(audio_file):
    # Initialize the recognizer
    r = sr.Recognizer()

    # Load audio file
    with sr.AudioFile(audio_file) as source:
        # Read the entire audio file
        audio = r.record(source)

    try:
        text = r.recognize_sphinx(audio)
        return text
    except sr.UnknownValueError:
        print("Speech recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Error with PocketSphinx recognition: {e}")


audio_file = "sample.wav"
text = convert_audio_to_text(audio_file)
print(text)
```
