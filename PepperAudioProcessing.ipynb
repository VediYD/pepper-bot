```python
def eye_color(ip, port, rgb=None, status="ListenOn"):
    if rgb is None:
        if status == "ListenOn":
            rgb = [246, 226, 127]
        elif status == "SpeechDetected":
            rgb = [226, 195, 145]
        elif status == "EndOfProcess":
            rgb = [155, 190, 199]

    leds = ALProxy("ALLeds", ip, port)
    names = [
        "Face/Led/Red/Left/0Deg/Actuator/Value",
        "Face/Led/Red/Left/90Deg/Actuator/Value",
        "Face/Led/Red/Left/180Deg/Actuator/Value",
        "Face/Led/Red/Left/270Deg/Actuator/Value"
    ]

    leds.createGroup("AllLeds")
    leds.fadeListRGB("AllLeds", rgb)
```

```python
import time
from naoqi import ALProxy


ROBOT_IP = '10.104.23.217'

# Creates a proxy on the speech-recognition module
asr = ALProxy("ALSpeechRecognition", ROBOT_IP, 9559)
memory = ALProxy("ALMemory", ROBOT_IP, 9559)

asr.setLanguage("English")

# asr.stop()
# Example: Adds "yes", "no" and "please" to the vocabulary (without wordspotting)
vocabulary = ["yes", "no", "please", "pepper"]

asr.setVocabulary(vocabulary, False)

# Start the speech recognition engine with user Test_ASR
asr.subscribe("Test_ASR")
print 'Speech recognition engine started'
# time.sleep(20)
for i in range(100):
    time.sleep(0.5)
    status = memory.getData("ALSpeechRecognition/Status")
    print(status)
    if status == "SpeechDetected":
        eye_color(status=status, ip=ROBOT_IP, port=9559)

asr.unsubscribe("Test_ASR")
```
